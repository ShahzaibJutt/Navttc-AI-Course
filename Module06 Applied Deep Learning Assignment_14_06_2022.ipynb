```python
import pandas as pd
wineData = pd.read_csv('winequality-red.csv',delimiter=';')
```

```python
wineData.duplicated().sum()
```

```python
wineData.isna().sum()
```

```python
wineData.drop_duplicates(inplace=True)
```

```python
x = wineData.iloc[:,:11]
y = wineData.iloc[:,11]
x
y
```

```python
wineData.columns
```

```python
import sklearn.preprocessing as ps
import numpy as np
scaler = ps.StandardScaler()
from sklearn.compose import make_column_transformer
transformer = make_column_transformer((scaler, ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
       'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
       'pH', 'sulphates', 'alcohol']))
px = transformer.fit_transform(x)
py = scaler.fit_transform(np.c_[y])
```

```python
px.shape
```

```python
import tensorflow as tf
from tensorflow.keras import layers
from tensorflow.keras import models
model = models.Sequential()
model.add(layers.Dense(100, input_dim=11 , activation='relu'))
model.add(layers.Dense(75 , activation='relu'))
model.add(layers.Dense(50 , activation='relu'))
model.add(layers.Dense(20 , activation='relu'))
model.add(layers.Dense(1))
model.summary()
```

```python
model.compile(loss='mean_squared_error', optimizer='adam', metrics=['mse'])
```

```python
from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y = train_test_split(px,py)
```

```python
model.fit(train_x,train_y, epochs=50)
```

```python
pred_uy=model.predict(test_x)
pred_y = scaler.inverse_transform(pred_uy)
tested_y = scaler.inverse_transform(test_y)
```

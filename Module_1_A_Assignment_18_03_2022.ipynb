```python
Name = []
RollNo = []
total_students = 0
while 1:
    print('1. Add Students\n2. List all students.\n3. Search students\n4. Remove Students\n5. Display Record\n6. Exit')
    choice = int(input())
    if choice == 1:
        n = input("Enter student name: ")
        r = input("Enter student roll no: ")
        Name.append(n)
        RollNo.append(r)
        total_students = total_students+1
    elif choice == 2:
        for i in range(0, len(Name)):
            print('Name: {}\nRoll No: {}'.format(Name[i],RollNo[i]))
    elif choice == 3:
        s = input("Enter student name: ")
        for i in Name:
            if s == i:
                print("Student found!")
            else:
                print("No student found with this name!")
    elif choice == 4:
        n = input("Enter student name to remove: ")
        for i in range(0,len(Name)):
            if n == Name[i]:
                del Name[i]
                del RollNo[i]
                total_students = total_students-1
                print("Student Removed")
    elif choice == 5:
        print("Total students: ", total_students)
    elif choice == 6:
        break
        
```

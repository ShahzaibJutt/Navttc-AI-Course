```python
#task 1
import pandas as pd
Final_Evaluation = pd.read_excel('Final Evaluation.xlsx',sheet_name='Final Evaluation', header=2, index_col=1)
Final_Evaluation.dropna(axis = 1, how ='all', inplace = True)
Final_Evaluation.head()
Final_Evaluation.tail()
```

```python
#task 2
Assignments = pd.read_excel('Final Evaluation.xlsx',sheet_name='Assignments', header=2, index_col=1,usecols="B:I,U")
Quiz = pd.read_excel('Final Evaluation.xlsx',sheet_name='Quiz', header=2, index_col=1,usecols="B:F,AA")
Projects = pd.read_excel('Final Evaluation.xlsx',sheet_name='Projects', header=2, index_col=1,usecols="B:F,AA")
Final_Result = pd.concat([Assignments, Quiz, Projects],axis=1)
Final_Result
```

```python
#task 3
Final_Result.fillna(0)
```

```python
#task 5
Result = Final_Evaluation["Total\n(100)"].tolist()
# pasfail = lambda x: True if x > 50 else False
# pasfail (Result)
passfail = ["Pass" if x > 50 else "Fail" for x in Result]
Final_Evaluation["Final_Result"] = passfail
Final_Evaluation
```

```python
Final_Evaluation["GRADE"] = Final_Evaluation["Roll No."]
for i in range(1,len(Final_Evaluation["GRADE"])+1):
    if 90.0 <= Final_Evaluation["Total\n(100)"][i] <= 100.0:
        Final_Evaluation["GRADE"][i]='A+'
    elif 80.0 <= Final_Evaluation["Total\n(100)"][i] < 90.0:
         Final_Evaluation["GRADE"][i]='A+'
    elif 70.0 <= Final_Evaluation["Total\n(100)"][i] < 80.0:
         Final_Evaluation["GRADE"][i]='B'
    elif 60.0 <= Final_Evaluation["Total\n(100)"][i] < 70.0:
         Final_Evaluation["GRADE"][i]='C'
    elif 50.0 <= Final_Evaluation["Total\n(100)"][i] < 60.0:
         Final_Evaluation["GRADE"][i]='D'
    elif 50 > Final_Evaluation["Total\n(100)"][i]:
         Final_Evaluation["GRADE"][i]='FAIL'
Final_Evaluation
```

```python
#task 7
Final_Evaluation = Final_Evaluation.sort_values(['Total\n(100)'],ascending=False)
Final_Evaluation
```

```python
#task 8
total_mean = Final_Evaluation['Total\n(100)'].mean()
print(total_mean)
```

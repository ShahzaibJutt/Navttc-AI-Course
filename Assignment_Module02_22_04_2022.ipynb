```python
# Reading CSV File
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv('Cat_human_New.csv')
humans = df.iloc[0:100]
cats = df.iloc[100:]
```

```python
# Relation between Weight 
plt.scatter(humans['Weight'].index,humans['Weight'],color='c')
plt.scatter(cats['Weight'].index,cats['Weight'],color='k')
plt.legend(['Human','Cats'])
plt.show()
```

```python
# Relation between Height
plt.scatter(humans['Height'].index,humans['Height'],color='c')
plt.scatter(cats['Height'].index,cats['Height'],color='k')
plt.legend(['Human','Cats'])
plt.show()
```

```python
# 'Relation between Height and Weight'
plt.scatter(humans['Height'].index,humans['Height'].values,color='c',marker='*')
plt.scatter(cats['Height'].index,cats['Height'].values,color='k',marker='*')
plt.legend(["Human","Cats"])
plt.scatter(humans['Weight'].index,humans['Weight'].values,color='c')
plt.scatter(cats['Weight'].index,cats['Weight'].values,color='k')
plt.show()
```

```python
# Relation between EyeColor
plt.scatter(humans['Eye_color'].index,humans['Eye_color'],color='c')
plt.scatter(cats['Eye_color'].index,cats['Eye_color'],color='k')
plt.legend(['Human','Cats'])
plt.show()
```

```python
# Relation between No of Legs
plt.scatter(humans['Legs'].index,humans['Legs'],color='c')
plt.scatter(cats['Legs'].index,cats['Legs'],color='k')
plt.legend(["Human","Cats"])
```

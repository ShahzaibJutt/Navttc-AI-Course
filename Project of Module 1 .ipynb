```python
def getPosInt(prompt):
    while 1:
        try:
            inp = input(prompt)
            if not inp:
                print('MISSING')
                continue
            if int(inp) < 0:
                raise ValueError
        except ValueError:
            if int(inp) < 0:
                print("{} is not greater than 0".format(inp))
                continue
        except:
            if inp.isdigit():
                print("{} is not an integar".format(inp))
                continue
            else:
                print('Is not valid!')
        else:
            return int(inp)
            break
def getBoundedInt(prompt, lowerBound, upperBound):
    while 1:
        try:
            inp = input(prompt)
            if not inp:
                print('MISSING')
                continue
            if int(inp) < lowerBound or int(inp) > upperBound:
                raise ValueError
        except ValueError:
            if int(inp) < lowerBound or int(inp) > upperBound:
                print("{} Plz enter any valid value between {}-{}".format(inp,lowerBound, upperBound))
                continue
        except:
            if inp.isdigit():
                print("{} is not an integar".format(inp))
                continue
            else:
                print('Is not valid!')
        else:
            return int(inp)
            break

def addTrashToCan(trashCans, can, counters):
    if counters[can] == len(trashCans):
        counters[can] -= 1
    if trashCans[counters[can]][can] == 0:
        trashCans[counters[can]][can] += (1+counters[can])
        if counters[can] != len(trashCans):
            counters[can] += 1     
        return False
    else:
        return True
            
def displayTrashCans(trashCans):
    for i in range(len(trashCans)-1,-1,-1):
        for j in range(0,len(trashCans[i])):
            print('| {} '.format(trashCans[i][j]), end=' ') if trashCans[i][j] != 0 else print('| ', end = ' ')
        print('\n')
    
if __name__ == '__main__':
    trashcans = getPosInt("Enter the number of trash cans (> 0): ")
    nobpt = getPosInt("Enter the number of bags per trash can (> 0): ")
    trashcan=[[0]*trashcans for _ in range(nobpt)]
    noofbags = [0]*trashcans
    while 1:
        can = getBoundedInt("Enter a trash can no between 0 and {}: ".format(trashcans-1),0,trashcans)
        res = addTrashToCan(trashcan,can,noofbags)
        displayTrashCans(trashcan)
        if res == True:
            print('Trash can {} is full! Time to have the trash picked up.'.format(can))
            displayTrashCans(trashcan)
            print('Programmed by Muhammad Shahzaib')
            break
```
